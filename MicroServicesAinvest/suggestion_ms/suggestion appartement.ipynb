```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('C:\\Users\\USER\\appartement73%.csv', sep=',')
pd.options.display.max_columns=None
dataset.head(10)
```

```python
def extract_year(date_str):
    return int(date_str.split("/")[-1])

# Appliquer la fonction à chaque valeur de la colonne "date" pour créer la colonne "year"
dataset["year"] = dataset["Date mutation"].apply(extract_year)
dataset
```

```python
dataset['Code departement'] = dataset['Code departement'].astype(str)
```

```python
dataset.shape
```

```python
commune = input("Enter a commune: ")
section = input("Enter a section: ")
surface_bati = float(input("Enter the surface reelle bati: "))
num_pieces = float(input("Enter the number of pieces principales: "))

filtered_data = dataset[(dataset['Commune'] == commune) &
                         (dataset['Section'] == section) &
                         (dataset['Surface reelle bati'] == surface_bati) &
                         (dataset['Nombre pieces principales'] == num_pieces)]
if filtered_data.empty:
    print("No data available for the inputs provided.")
else:
    model = LinearRegression()

    features = ['year']
    target = 'Valeur fonciere'
    model.fit(filtered_data[features], filtered_data[target])

    future_years = pd.Series(np.arange(2023, 2028))
    future_data = pd.DataFrame({'Surface reelle du bati': [surface_bati]*5, 'Nombre pieces principales': [num_pieces]*5, 'Commune': [commune]*5, 'Section': [section]*5, 'year': future_years})
    predictions = model.predict(future_data[features])

    lowest_price_year = None
    lowest_price = float('inf')
    highest_price_year = None
    highest_price = float('-inf')
    for i in range(len(predictions)):
        if predictions[i] < lowest_price:
            lowest_price_year = future_years[i]
            lowest_price = predictions[i]
        if predictions[i] > highest_price:
            highest_price_year = future_years[i]
            highest_price = predictions[i]

    price_diff = ((highest_price - lowest_price) / lowest_price) * 100

    if lowest_price_year is not None and highest_price_year is not None:
        print("Vous pouvez investir votre budget en:", lowest_price_year,"et revendre votre immobilier en:", highest_price_year,"avec un gain de : {:.2f}%".format(price_diff))
```
